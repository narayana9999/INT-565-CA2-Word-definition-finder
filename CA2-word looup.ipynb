# Implement a simple Feed Forward NN, with back Propgation on IRIS dataset and caluculate overall loss and accuracy 

```python
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import *
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.losses import sparse_categorical_crossentropy
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = load_iris()
data = dataset['data']
label = dataset['target']
```

```python
label_catagotical = to_categorical(label)
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,label,test_size=0.2,random_state=42)
```

```python
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
model_mlp = MLPClassifier(hidden_layer_sizes=(512,),activation="relu",max_iter=1000)
model_mlp.fit(x_train_scaled,y_train)
```

```python
model_mlp.score(x_train_scaled,y_train)
```

```python
train_pred = model_mlp.predict(x_train_scaled)
train_accuracy = accuracy_score(y_train,train_pred)
print(f"Train accuracy :{train_accuracy}")
print(f"Train loss : {1 - train_accuracy}")
# train_accuracy, (1-train_accuracy)
```

```python
test_pred =model_mlp.predict(x_test_scaled)
test_accuracy = accuracy_score(y_test,test_pred)
print(f"Test accuracy :{test_accuracy}")
print(f"Test loss : {1 - test_accuracy}")
# test_accuracy,(1-test_accuracy)
```
